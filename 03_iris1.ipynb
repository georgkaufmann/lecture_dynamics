The iris flowers have **two sets of leaves**, the **sepal** leaves (covering the buds)
and the **petal** leaves (the flowering leaves).

The length and width of these two leaves are characteristic for the different iris species.

<img style="height: 300px;" src="images/sepal_petal_leaves.jpg">
Our idea is as follows:
- Measure length and width of sepal and petal leaves of enough iris flowers (of the three types) to
obtain a set of observations $o_i$.
- The length and width of sepal and petal leaves are our parameter values $a_j$.

For the three named species, 150 samples are recorded in the `iris data set`.
