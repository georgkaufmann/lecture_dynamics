```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap 
import csv
```

```python
n=150
m=4
X = np.zeros([n*m]).reshape(n,m)
y = np.zeros([n])

csvfile = open('data/iris.csv', newline='')
iris = csv.reader(csvfile, delimiter=',')
next(iris)
i=0
for row in iris:
    X[i,0] = row[0]
    X[i,1] = row[1]
    X[i,2] = row[2]
    X[i,3] = row[3]
    if (row[4]=='Iris-setosa'):
        y[i] = 0
    elif (row[4]=='Iris-versicolor'):
        y[i] = 1
    elif (row[4]=='Iris-virginica'):
        y[i] = 2
    i=i+1
print(X.ndim,X.shape,X[0])
print(y.ndim,y.shape,y[0])
```

```python
from sklearn import neighbors
clf = neighbors.KNeighborsClassifier(n_neighbors=10)
```

```python
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
h=0.01
#--------0,1------------
#pair=[0,1]
Xpair = X[:, [0,1]]
clf.fit(Xpair, y)
x_min, x_max = Xpair[:, 0].min() - 1, Xpair[:, 0].max() + 1
y_min, y_max = Xpair[:, 1].min() - 1, Xpair[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                    np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
fig, axs = plt.subplots(2, 3, figsize=(14,7), sharey=False)
axs[0].pcolormesh(xx, yy, Z, cmap=cmap_light, alpha=0.5)
axs[0].scatter(X[:,0], X[:,1], c=y, cmap=cmap_bold, s=80, marker='o')
axs[0].set_xlim(xx.min(), xx.max())
axs[0].set_ylim(yy.min(), yy.max())
```
